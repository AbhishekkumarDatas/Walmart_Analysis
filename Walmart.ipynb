```python
df['Unemployment'].corr(df['Weekly_Sales'])
```

```python
def calc_corr(group):
    return group['Unemployment'].corr(group['Weekly_Sales'])

correlations = df.groupby('Store').apply(calc_corr).reset_index()
correlations.columns = ['Store', 'Unemployment_Sales_Correlation']
sorted_correlations = correlations.sort_values(by='Unemployment_Sales_Correlation')
print(sorted_correlations)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Store', y='Unemployment_Sales_Correlation', data=sorted_correlations)
plt.title('Correlation between Unemployment and Weekly Sales by Store')
plt.xlabel('Store')
plt.ylabel('Correlation Coefficient')
plt.show()
```
