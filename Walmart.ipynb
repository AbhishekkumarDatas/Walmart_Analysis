# Q If the Weekly Sales show a seasonal trend, when and what could be the reson?

```python
def season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return None 
df['season'] = df['Months'].apply(season)
```

```python
df.groupby(['season'],as_index= False)['Weekly_Sales'].sum().sort_values(by='Weekly_Sales',ascending=False)
```

```python
# Pearson Correlation

# The Pearson Correlation measures the linear dependence between two variables X and Y.
# # 
# The resulting coefficient is a value between -1 and 1 inclusive, where:

# 1: Perfect positive linear correlation. 0: No linear correlation, the two variables most likely do not affect each other.
# -1: Perfect negative linear correlation.

# By convention, when the

# p-value is < 0.001: we say there is strong evidence that the correlation is significant.
# p-value is < 0.05: there is moderate evidence that the correlation is significant.
# p-value is < 0.1: there is weak evidence that the correlation is significant.
# p-value is > 0.1: there is no evidence that the correlation is significant.
```

```python
from scipy.stats import pearsonr
pearsonr_coeff, p_value =pearsonr(df['Fuel_Price'], df['Weekly_Sales'])
print(pearsonr_coeff, p_value)
```

```python
plt.figure(figsize=(12,8))
sns.regplot(x='Fuel_Price', y ='Weekly_Sales',data =df)
```

```python
# Conclusion:

# Since the p-value is > 0.1, the correlation between fuel price and weekly sales is not statistically significant.
# Fuel price does not seem like a good predictor of the weekly sales at all since the regression line is close to horizontal.
# Therefore, it's not a reliable variable
```

```python
# Unemployment vs. Weekly Sales
pearsonr_coeff, p_value =pearsonr(df['Unemployment'], df['Weekly_Sales'])
print(pearsonr_coeff, p_value)
```

```python
plt.figure(figsize=(12,8))
sns.regplot(x='Unemployment', y ='Weekly_Sales',data =df)
```

```python
# Conclusion:-

# Since the p-value is < 0.001, the correlation between unemployment and weekly sales is strong evidence that the correlation is significant.
# Unemployment seems like a good predictor of the weekly sales, The higher the unemployment rate, the lower the weekly sales
```

```python
Q1,Q3= df_new['Unemployment'].quantile([.25,.75])
IQR= Q3-Q1
LL=Q1-1.5*(IQR)
UL=Q3+1.5*(IQR)
```

```python
def limit_imputer(x):
    if x>UL:
        return UL
    if x<LL:
        return LL
    else:
        return x
        
```

```python
df_new['Unemployment'] =df_new['Unemployment'].apply(limit_imputer)
```

```python
Q1,Q3= df_new['Temperature'].quantile([.25,.75])
IQR= Q3-Q1
LL=Q1-1.5*(IQR)
UL=Q3+1.5*(IQR)

def limit_imputer(x):
    if x>UL:
        return UL
    if x<LL:
        return LL
    else:
        return x
        
```

```python
df_new['Temperature']=df_new['Temperature'].apply(limit_imputer)
```
